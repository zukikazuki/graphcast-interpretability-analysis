```python
import xarray as xr
from pathlib import Path

# 予測出力の親ディレクトリ（forecastsまで）
forecasts_dir = Path("/media/dl-box/SSD-SCTU3A/graphcast_data/outputs/forecasts")

# 対象ファイル名（GNNステップ16の例）
file_name = "source-era5_date-2022-01-01T12_res-1.0_levels-13_steps-01_gnn16_predictions.nc"
path = forecasts_dir / file_name

# データセットを開いて構造を確認
with xr.open_dataset(path) as ds:
    print(ds)
    print("\ncoords:")
    for k, v in ds.coords.items():
        print(f"- {k}: shape={v.shape}, attrs={dict(v.attrs)}")
    print("\ndata_vars:")
    for k, v in ds.data_vars.items():
        print(f"- {k}: shape={v.shape}, dims={v.dims}, attrs={dict(v.attrs)}")

```
